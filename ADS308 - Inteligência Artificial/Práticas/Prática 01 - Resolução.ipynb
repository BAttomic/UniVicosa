## 1. Estruturas Base: Listas

Listas são coleções ordenadas e mutáveis de itens. Em dados, muitas vezes representam uma coluna ou uma série de observações.

```python
# Criando uma lista
idades = [22, 34, 45, 19, 28]

# Acessando elementos (índice começa em 0)
primeiro_cliente = idades[0]
print("Primeiro cliente:", primeiro_cliente)

# Adicionando um novo dado
idades.append(31)
print("Lista atualizada:", idades)
```

## 2. Listas: Fatiamento (Slicing)

Fatiamento é crucial para dividir datasets no futuro (ex: separar treino e teste).

```python
dados = [10, 20, 30, 40, 50, 60]

# Pegando do início até o índice 2 (exclusivo)
primeiros_tres = dados[:3]  # [10, 20, 30]
print("Primeiros três:", primeiros_tres)

# Pegando do índice 3 até o final
resto = dados[3:]  # [40, 50, 60]
print("Resto:", resto)

# Pegando o último elemento
ultimo = dados[-1]  # 60
print("Último:", ultimo)
```

## 3. Estruturas Base: Dicionários

Dicionários são coleções de pares chave-valor. Eles são a base para entender o formato JSON e como construir DataFrames no Pandas.

```python
# Criando um dicionário
cliente = {
    "nome": "Ana",
    "idade": 28,
    "comprou": True
}

# Acessando valor através da chave
idade_cliente = cliente["idade"]
print("Idade do cliente:", idade_cliente)
```

## 4. Dicionários: Iteração

```python
vendas = {"jan": 100, "fev": 150, "mar": 200}

# Iterando sobre chaves e valores
for mes, valor in vendas.items():
    print(f"Mês: {mes} - Vendas: {valor}")

# Adicionando nova chave-valor
vendas["abr"] = 180
print("\nVendas atualizadas:", vendas)
```
