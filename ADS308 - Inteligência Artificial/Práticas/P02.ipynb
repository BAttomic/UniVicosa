## 3. Função Auxiliar: Vizinhos Válidos

Retorna os vizinhos (cima, baixo, esquerda, direita) que estão dentro do mapa e não são obstáculos.

```python
def vizinhos_validos(pos, mapa):
    linhas, colunas = mapa.shape
    x, y = pos
    vizinhos = []
    for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
        nx, ny = x + dx, y + dy
        if 0 <= nx < linhas and 0 <= ny < colunas and mapa[nx][ny] == 0:
            vizinhos.append((nx, ny))
    return vizinhos
```

## 4. Função de Visualização

- **Preto:** Paredes/Obstáculos
- **Vermelho:** Caminho final encontrado
- **Azul Claro:** Nós visitados (explorados)

```python
def visualizar(mapa, visitados, caminho, titulo="Pathfinding"):
    visual = np.copy(mapa).astype(float)
    # 0 = livre (branco), 1 = parede (preto)
    # 2 = visitado (azul claro), 3 = caminho (vermelho)
    for pos in visitados:
        if visual[pos] == 0:
            visual[pos] = 2
    for pos in caminho:
        visual[pos] = 3

    cmap = ListedColormap(["white", "black", "lightblue", "red"])
    plt.figure(figsize=(6, 6))
    plt.imshow(visual, cmap=cmap, vmin=0, vmax=3)
    plt.title(f"{titulo}\nNós explorados: {len(visitados)} | Caminho: {len(caminho)} passos")
    plt.xticks(range(mapa.shape[1]))
    plt.yticks(range(mapa.shape[0]))
    plt.grid(True, linewidth=2)
    plt.show()
```

## 5. Função para Reconstruir o Caminho

A partir do dicionário de pais, reconstrói o caminho do objetivo até o início.

```python
def reconstruir_caminho(pai, inicio, objetivo):
    caminho = []
    atual = objetivo
    while atual != inicio:
        caminho.append(atual)
        atual = pai[atual]
    caminho.append(inicio)
    caminho.reverse()
    return caminho
```
